```python
#导入数据
import numpy as np
from numpy import *
import pandas as pd

txt_path = '../data/browsing.txt'
f = open(txt_path)
data_lists = f.readlines() 

dataset= []
for data in data_lists:
    data1 = data.strip('\n') 
    data2 = data1.split() 
    dataset.append(data2) 

dataset = np.array(dataset)

#去除重复数据
for i in range(len(dataset)):
    dataset[i] = list(set(dataset[i]))
print(dataset)
```

d. 

```python
# 1）扫描整个数据集，得到所有出现过的item，作为候选频繁1项集,并记录每个item出现的次数
count1 = {}
for i in dataset:
    for j in i:
        count1[j] = count1.get(j, 0)+1

# 2）筛选出现次数大于100的item,并将出现次数少于100次的item进行剪枝，得到真正的频繁1项集
cut_count1 = {k:v for k, v in count1.items() if v >= 100}
cut_dataset = [[j for j in i if j in cut_count1] for i in dataset]
num_freSeq1=list(cut_count1.keys())
print("frequent items after 1st:",len(num_freSeq1)) 
print('经核验准确')

# 3）生成频繁2项集
count2 = {}
items = cut_count1.keys()
for i in range(len(cut_dataset)):
    for j in range(len(cut_dataset[i])):
        for k in range(j+1, len(cut_dataset[i])):
            pair = tuple(sorted([cut_dataset[i][j],cut_dataset[i][k]]))
            count2[pair] = count2.get(pair,0)+1
cut_count2 = {k:v for k, v in count2.items() if v >= 100} #剪枝生成真正的频繁2项集

# 4) 计算confidence,并列出top5

twoPairs = pd.DataFrame({'pair':cut_count2.keys(), 'count':cut_count2.values()})
twoPairs = twoPairs.sort_values(by = 'count', ascending = False)
twoPairs['countX'] =  twoPairs['pair'].apply(lambda x: cut_count1[x[0]])
twoPairs['countY'] =  twoPairs['pair'].apply(lambda x: cut_count1[x[1]])
twoPairs['confidenceX->Y'] = twoPairs['count']/twoPairs['countX']
twoPairs['confidenceY->X'] = twoPairs['count']/twoPairs['countY']

print('----------------------------------------------------------------------')
print('最终结果为：')
print('\n')
print('support大于100的频繁2项集：')
print(twoPairs['pair'])
print('\n')
print('the top5 confidence:')
print(twoPairs.sort_values(by = 'confidenceX->Y', ascending = False).iloc[:5,:])
print(twoPairs.sort_values(by = 'confidenceY->X', ascending = False).iloc[:5,:])

```

```python
count3={}
# 生成频繁3项集
for i in range(len(cut_dataset)):
    temp = cut_dataset[i]
    for j in range(len(temp)):
        for k in range(j+1, len(temp)):
            if tuple(sorted((temp[j],temp[k]))) in cut_count2:
                for m in range(k+1, len(temp)):
                    pair = tuple(sorted((temp[j],temp[k],temp[m])))
                    count3[pair] = count3.get(pair,0) + 1

#筛选出现次数大于100的item,并将出现次数少于100次的item进行剪枝，得到真正的频繁3项集                   
cut_count3 = {k:v for k, v in count3.items() if v >= 100}              

threePairs = pd.DataFrame({'pair':cut_count3.keys(), 'count':cut_count3.values()})
threePairs = threePairs.sort_values(by = 'count', ascending = False)

threePairs['countXY'] = threePairs['pair'].apply(lambda x: cut_count2[tuple(sorted((x[0],x[1])))])
threePairs['countYZ'] = threePairs['pair'].apply(lambda x: cut_count2[tuple(sorted((x[1],x[2])))])
threePairs['countXZ'] = threePairs['pair'].apply(lambda x: cut_count2[tuple(sorted((x[0],x[2])))])

threePairs['confidenceXY->Z'] = threePairs['count']/threePairs['countXY']
threePairs['confidenceYZ->X'] = threePairs['count']/threePairs['countYZ']
threePairs['confidenceXZ->Y'] = threePairs['count']/threePairs['countXZ']




print('----------------------------------------------------------------------')
print('最终结果为：')
print('\n')
print('support大于100的频繁3项集：')
print(threePairs['pair'])
print('\n')
print('the top5 confidence:')
print(threePairs.sort_values(by = 'confidenceXY->Z', ascending = False).iloc[:5,:])
print(threePairs.sort_values(by = 'confidenceYZ->X', ascending = False).iloc[:5,:])
print(threePairs.sort_values(by = 'confidenceXZ->Y', ascending = False).iloc[:5,:])

```
